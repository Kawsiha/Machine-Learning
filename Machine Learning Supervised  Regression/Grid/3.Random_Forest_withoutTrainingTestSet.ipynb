```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
para_grid={'criterion':('squared_error', 'absolute_error', 'friedman_mse', 'poisson'),'max_features':('sqrt', 'log2') }
grid=GridSearchCV(RandomForestRegressor(),para_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("Best Result:{}" .format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age=int(input("Age:"))
bmi=float(input("bmi:"))
children=int(input("Children:"))
sex_male=int(input("Sex_male 0 or 1:"))
smoker_yes=int(input("smoker_yes 0 or 1:"))
```

```python
Future_pred=grid.predict([[age,bmi,children,sex_male,smoker_yes]])
print("Future_Prediction:{}" .format(Future_pred))
```
