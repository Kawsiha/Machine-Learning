```python
#1. Why does this feature selection with chi2 fail? Chi2 should have non-negative value
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler

X = [[-1, 2, 3], [4, 0, -6], [7, 8, 9], [1, 2, 3]]
y = [0, 1, 0, 1]

sc=MinMaxScaler()
x=sc.fit_transform(X)
selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(x, y)
X_new
```

```python
#6.
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

items = ["action movie", "romantic comedy", "sci-fi thriller"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(items)

similarity = cosine_similarity(X[0:1], X[1:2])
similarity

```
