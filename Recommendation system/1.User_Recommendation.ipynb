```python
import pandas as pd
import numpy as np
```

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
ratings
```

```python
n_users=ratings.user_id.unique().shape[0]
n_movies=ratings.movie_id.unique().shape[0]

print("No. of Unique Users:",n_users)
print("No. of Unique Movies:",n_movies)
```

```python
#Creating Pivot table for User and Movies

datama=ratings.pivot_table(index='user_id',columns='movie_id',values='rating')
datama
```

```python
data_matrix=datama.fillna(0)
data_matrix
```

```python
#Find Cosine Similatity for user and Item

from sklearn.metrics.pairwise import pairwise_distances 
user_similarity = pairwise_distances(data_matrix, metric='cosine')
item_similarity = pairwise_distances(data_matrix.T, metric='cosine')
```

```python
#Using formula for user and item we are calcuating the score value

def predict(ratings, user_similarity,item_similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        mean_user_rating = np.array(mean_user_rating)
        #We use np.newaxis so that mean_user_rating has same format as ratings
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + user_similarity.dot(ratings_diff) / np.array([np.abs(user_similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(item_similarity) / np.array([np.abs(item_similarity).sum(axis=1)])
    return pred
```

```python
# prediction Table
user_prediction = predict(data_matrix,user_similarity, item_similarity,type='user')
item_prediction = predict(data_matrix, user_similarity,item_similarity, type='item')
```

```python
def UserBased(input_user,user_similarity,user_prediction,sim_user_count,thres):

    user_sim_table=pd.DataFrame(user_similarity)
    
    sim_input_user = user_sim_table[input_user].sort_values(ascending=True).head(sim_user_count).index

    sim_user_input=list(sim_input_user)

    sim_user_movieid=[]
    for sim in sim_user_input:
        sim1= list (ratings[ratings.user_id == sim]["movie_id"])
        sim_user_movieid.append(sim1)

    sim_user_movieid_single= [x for sub in sim_user_movieid for x in sub]

    unique_movieid_sim_user= set(sim_user_movieid_single)

    input_user_watched_movies=list (ratings[ratings.user_id == input_user]["movie_id"].values)
#Recommendation movies for User
    rec=[]
    for i in unique_movieid_sim_user:
        if (i in input_user_watched_movies):
            pass
        else:
            rec.append(i)
            
#From recommendation list selecting only highest rated(predicted) value
    high_rated=[]
    user_pred_table=pd.DataFrame(user_prediction)
    user_pred_trans=user_pred_table.T
    input_user_pre=pd.DataFrame(user_pred_trans[input_user])
    input_user_pred=input_user_pre.T

    for re in rec:
        value=input_user_pred[re].values
        if (value >= thres):
            high_rated.append(re)

    i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
    
    items = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
    
#Creating Movie list from movie id
    movie_title=[]
    for movieid in high_rated:
        mov= items[items["movie id"] == movieid]["movie title"].values
        movie_title.append(mov)

    movie_title_list=[]
    for m in movie_title:
        mv=list(m)
        movie_title_list.append(mv)
        
    Final_Rec_Movies= [j for su in movie_title_list for j in su]
    
    print("The common Movie in Recom & User:",list(set(rec)&set(input_user_watched_movies)))
    print("Recommended Movies:")
    for movie in Final_Rec_Movies:
        print(movie)

    return Final_Rec_Movies
```
