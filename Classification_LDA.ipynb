```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#train_test split
def train_test(Indep, dep):
    x_train,x_test,y_train,y_test=train_test_split(Indep, dep, test_size=0.25, random_state=42)
    sc = StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)
    return x_train,x_test, y_train,y_test 
```

```python
#cm score
def CM(classifier,x_test):
    y_pred=classifier.predict(x_test)
    cm=confusion_matrix(y_test,y_pred)
    cr=classification_report(y_test,y_pred)
    Accuracy=accuracy_score(y_test, y_pred )
    return cm, cr, Accuracy, x_test, y_test
```

```python
#feature Selection
def lda(Indep,dep,n_components):
    model = LinearDiscriminantAnalysis(n_components=n_components)
    feature= model.fit_transform(Indep,dep)
    return feature
```

```python
#RandomForestClassifier Model
def rf(x_train,y_train,x_test):
    classifier = RandomForestClassifier()
    classifier.fit(x_train,y_train)
    cm, cr, Accuracy, x_test, y_test=CM(classifier,x_test)
    return cm, cr, Accuracy, x_test, y_test
```

```python
#GaussianNB Model
def gaus(x_train,y_train,x_test):
    classifier=GaussianNB()
    classifier.fit(x_train,y_train)
    cm, cr, Accuracy, x_test, y_test=CM(classifier,x_test)
    return cm, cr, Accuracy, x_test, y_test
```

```python
#KNeighborsClassifier Model
def KN(x_train,y_train,x_test):
    classifier=KNeighborsClassifier()
    classifier.fit(x_train,y_train)
    cm, cr, Accuracy, x_test, y_test=CM(classifier,x_test)
    return cm, cr, Accuracy, x_test, y_test
```

```python
#DecisionTreeClassifier Model
def DT(x_train,y_train,x_test):
    classifier=DecisionTreeClassifier()
    classifier.fit(x_train,y_train)
    cm, cr, Accuracy, x_test, y_test=CM(classifier,x_test)
    return cm, cr, Accuracy, x_test, y_test
```

```python
#SVC Model
def SVCM(x_train,y_train,x_test):
    classifier=SVC()
    classifier.fit(x_train,y_train)
    cm, cr, Accuracy, x_test, y_test=CM(classifier,x_test)
    return cm, cr, Accuracy, x_test, y_test
```

```python
#Table Creation
def lda_classifier(rfm, guasm, kn, dtm, svcm):
    dataframe = pd.DataFrame(index=['LDA'],columns=['RandomForest', 'Guassian','KNeighbors','DecisionTree', 'SVC'])
    for num, idex in enumerate(dataframe.index):
        dataframe['RandomForest'][idex]=rfm[num]
        dataframe['Guassian'][idex]=guasm[num]
        dataframe['KNeighbors'][idex]=kn[num]
        dataframe['DecisionTree'][idex]=dtm[num]
        dataframe['SVC'][idex]=svcm[num]
    return dataframe
```

```python
dataset= pd.read_csv("Wine.csv")
ds=dataset
```

```python
ds= pd.get_dummies(ds, drop_first=True)
ds.shape
```

```python
Indep=ds.iloc[:, 0:13].values
dep=ds.iloc[:, 13].values
print(Indep.shape)
print(dep.shape)
```

```python
feature=lda(Indep,dep,5)
```

```python
rfm=[]
guasm=[]
kn=[]
dtm=[]
svcm=[]
```

```python
x_train,x_test, y_train,y_test= train_test(feature, dep)

cm, cr, Accuracy, x_test, y_test= rf(x_train,y_train,x_test)
rfm.append(Accuracy)

cm, cr, Accuracy, x_test, y_test= gaus(x_train,y_train,x_test)
guasm.append(Accuracy)

cm, cr, Accuracy, x_test, y_test= KN(x_train,y_train,x_test)
kn.append(Accuracy)

cm, cr, Accuracy, x_test, y_test= DT(x_train,y_train,x_test)
dtm.append(Accuracy)

cm, cr, Accuracy, x_test, y_test= SVCM(x_train,y_train,x_test)
svcm.append(Accuracy)

result = lda_classifier(rfm, guasm, kn, dtm, svcm)
```
